# **EDA**

It's time to create some data visualizations!

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Room.Board', y='Grad.Rate', hue='Private', palette='coolwarm')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Outstate', y='F.Undergrad', hue='Private', palette='coolwarm')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize=(8, 6))
g = sns.FacetGrid(data=df, hue='Private', height=6, aspect=1.5)
g.map(plt.hist, 'Outstate', bins=30, alpha=0.5, stacked=True)
g.add_legend()
plt.title('Out-of-State Tuition by Private School Status')
plt.xlabel('Out-of-State Tuition')
plt.ylabel('Count')
plt.show()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize=(8, 6))
g = sns.FacetGrid(data=df, hue='Private', height=6, aspect=1.5)
g.map(plt.hist, 'Grad.Rate', bins=30, alpha=0.5, stacked=True)
plt.title('Grad.Rate by Private School Status')
plt.xlabel('Grad.Rate')
plt.ylabel('Count')
plt.show()
```

# **K Means Cluster Creation**

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, n_init=10)
```

```python
kmeans.fit(df.drop('Private', axis=1))
```

```python
kmeans.cluster_centers_
```

# **Evaluation**

```python
df['Cluster'] = df['Private'].apply(lambda x: 1 if x=='Yes' else '0')
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
np.array(df['Cluster'].apply(lambda x: int(x)))
```

```python
print(confusion_matrix(np.array(df['Cluster'].apply(lambda x: int(x))), kmeans.labels_), classification_report(np.array(df['Cluster'].apply(lambda x: int(x))), kmeans.labels_), sep='\n')
```
